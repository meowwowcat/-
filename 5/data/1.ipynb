```python
import pandas as pd

filename = "noise.csv"

df = pd.read_csv(filename,skipfooter=11,usecols=[0,1],names=['time','voltage'])
```

```python
df
```

```python
data = df['voltage'].values
```

```python
import numpy as np

data_num = len(data)
data_min = min(data)
data_max = max(data)
data_ave = np.average(data)
data_sig = np.std(data)
data_usig = np.std(data, ddof = 1)

print(f"Number of data = {data_num}")
print(f"Maximum value = {data_max}")
print(f"Minimum value = {data_min}")
print(f"Average value of data = {data_ave}")
print(f"Standerd Devitation = {data_sig}")
print(f"Unbiased Standard Devitation = {data_usig}")
```

```python
import matplotlib.pyplot as plt

x_data = np.arange(1, data_num+1, 1)

plt.plot(x_data, data, color='k')
plt.xlabel('Time Series data', fontsize=14)
plt.ylabel('Voltage / V', fontsize=14)
plt.legend(['Voltage values'], fontsize=14)

plt.savefig('figure/fig1.png')
```

```python
import matplotlib.pyplot as plt

n, bins, _ = plt.hist(data, bins=10)
plt.xlabel('Voltage/V', fontsize=14)
plt.ylabel('Count', fontsize = 14)

plt.savefig('figure/fig2.png')
```
